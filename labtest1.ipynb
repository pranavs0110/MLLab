```python
# Importing Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# Importing Dataset
data = pd.read_csv("Indonesia_airport.csv")
```

```python
#overview of data
print(data.head(5))

print(data.info())

print(data.describe())
```

```python
#fill in missing values
data.fillna(data.mean(numeric_only=True) inplace=True)
print(data.isnull().sum)
```

```python
# Histogram
sns.histplot(x='Location Level', data=data, bins=20, kde=True, color='skyblue')
plt.title("location level")
plt.xlabel("Location Level)")
plt.ylabel("Frequency")
plt.show()

#More provinces given in dataset
```

```python
#count plot
sns.countplot(x='Province', data=data, palette='pastel')
plt.title("Distribution of provinces")
plt.xlabel("provinces")
plt.ylabel("Count")
plt.show()

#DKI jakarta is the most occuring province
```

```python
# Scatter Plot
sns.scatterplot(x='Location', y='Total Cases',  data=data, palette='coolwarm')
plt.title("cases based on location")
plt.xlabel("location")
plt.ylabel("cases")
plt.show()

#More cases in jakarta
```

```python
# Heatmap
numeric_data = data.select_dtypes(include=['number'])

sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, cbar=True)
plt.title("Correlation Heatmap")
plt.show()


```

```python
# Boxplot
sns.boxplot(x='Location Level', y='New Cases', data=data, palette='muted')
plt.title("Location level in relation to new cases")
plt.xlabel("location level")
plt.ylabel("cases")
plt.show()

#although there are more provinces country has more occurrences of cases
```

```python
# Violin Plot
sns.violinplot(x='Island', y='Population', data=data, palette='pastel')
plt.title("islands and population")
plt.xlabel("island")
plt.ylabel("population")
plt.show()

#jawa has the most population of any island
```

```python
sns.barplot(x='Island', y='Total Regencies', data=data, palette='pastel')
plt.title("islands and regencies")
plt.xlabel("island")
plt.ylabel("regencies")
plt.show()
```
